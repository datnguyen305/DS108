## a) Chức năng (nội dung) của bộ dữ liệu là gì?


**Đáp Án**: Nội dung của bộ dữ liệu là 1 thí nghiệm khảo sát trên 30 tình nguyện viên ở độ tuổi từ 19 - 48 tuổi. Mỗi người thực hiện 6 động tác, đi bộ, leo cầu thang, xuống cầu thang, ngồi, đứng và nằm, và trên eo mang theo điện thoại. Dùng cảm bieense gia tốc và con quay hồi chuyển để bắt được lần lượt 3 trục gia tốc và 3 trục góc ở mức 50Hz. Bộ dữ liệu được label bằng tay. Bộ dataset được chia 70/30. Các tín hiệu được thêm noise và chia thành những đoạn 2.56 giây và 50% chồng nhau. Tín hiệu của cảm biến gia tốc, bao gồm chỉ số trọng lực và chuyển động của cơ thể, được tách ra nhờ Butterworth. Chỉ số trọng lực được xem là chỉ số có tần số thấp, giảm đi 0.3Hz. Mỗi đoạn sẽ có 1 vector chứa những thuộc tính được tính từ những biến từ miền thời gian và tần suất.

**Trích**: "The experiments have been carried out with a group of 30 volunteers within an age bracket of 19-48 years. Each person performed six activities (WALKING, WALKING_UPSTAIRS, WALKING_DOWNSTAIRS, SITTING, STANDING, LAYING) wearing a smartphone (Samsung Galaxy S II) on the waist. Using its embedded accelerometer and gyroscope, we captured 3-axial linear acceleration and 3-axial angular velocity at a constant rate of 50Hz. The experiments have been video-recorded to label the data manually. The obtained dataset has been randomly partitioned into two sets, where 70% of the volunteers was selected for generating the training data and 30% the test data.
The sensor signals (accelerometer and gyroscope) were pre-processed by applying noise filters and then sampled in fixed-width sliding windows of 2.56 sec and 50% overlap (128 readings/window). The sensor acceleration signal, which has gravitational and body motion components, was separated using a Butterworth low-pass filter into body acceleration and gravity. The gravitational force is assumed to have only low frequency components, therefore a filter with 0.3 Hz cutoff frequency was used. From each window, a vector of features was obtained by calculating variables from the time and frequency domain. See 'features_info.txt' for more details."

- 'activity_labels.txt': Label tương ứng với hành động.

- 'train/subject_train.txt': mỗi dòng định danh ai là người thực hiện hành động, đánh dấu từ 1 - 30. 

- 'train/Inertial Signals/total_acc_x_train.txt': Tín hiệu gia tốc, chuẩn hóa theo đơn vị 'g'. Mỗi dòng có 128 vector. Tương tự cho 'total_acc_x_train.txt' and 'total_acc_z_train.txt' ở trục Y và Z. 

- 'train/Inertial Signals/body_acc_x_train.txt': Tín hiệu gia tốc được tính bằng cách tính hiệu giữa total_acc với chỉ số trọng lực. 

- 'train/Inertial Signals/body_gyro_x_train.txt': Chỉ số góc được tính bằng radient. 

**Trích**: - 'README.txt'

- 'features_info.txt': Shows information about the variables used on the feature vector.

- 'features.txt': List of all features.

- 'activity_labels.txt': Links the class labels with their activity name.

- 'train/X_train.txt': Training set.

- 'train/y_train.txt': Training labels.

- 'test/X_test.txt': Test set.

- 'test/y_test.txt': Test labels.

The following files are available for the train and test data. Their descriptions are equivalent. 

- 'train/subject_train.txt': Each row identifies the subject who performed the activity for each window sample. Its range is from 1 to 30. 

- 'train/Inertial Signals/total_acc_x_train.txt': The acceleration signal from the smartphone accelerometer X axis in standard gravity units 'g'. Every row shows a 128 element vector. The same description applies for the 'total_acc_x_train.txt' and 'total_acc_z_train.txt' files for the Y and Z axis. 

- 'train/Inertial Signals/body_acc_x_train.txt': The body acceleration signal obtained by subtracting the gravity from the total acceleration. 

- 'train/Inertial Signals/body_gyro_x_train.txt': The angular velocity vector measured by the gyroscope for each window sample. The units are radians/second. 


## d) Các thông tin nào mà các record lưu trữ?

# 3) Merge bộ dữ liệu train và test lại với nhau.


```python
import pandas as pd 
```

```python
# Đọc file train
X_train = pd.read_csv('getdata_projectfiles_UCI HAR Dataset/UCI HAR Dataset/train/X_train.txt', delim_whitespace=True, header=None)
y_train = pd.read_csv('getdata_projectfiles_UCI HAR Dataset/UCI HAR Dataset/train/y_train.txt', delim_whitespace=True, header=None)

print(X_train.shape, y_train.shape)
```

```python
# Đọc file test
X_test = pd.read_csv('getdata_projectfiles_UCI HAR Dataset/UCI HAR Dataset/test/X_test.txt', delim_whitespace=True, header=None)
y_test = pd.read_csv('getdata_projectfiles_UCI HAR Dataset/UCI HAR Dataset/test/y_test.txt', delim_whitespace=True, header=None)

print(X_test.shape, y_test.shape)
```

```python
# Đọc file features
features = pd.read_csv('getdata_projectfiles_UCI HAR Dataset/UCI HAR Dataset/features.txt', delim_whitespace=True, header=None)
features.columns = ['index', 'feature']
features = features['feature'].tolist()
# Lấy tên các cột từ file features và đặt cho X_train và X_test
X_train.columns = features
X_test.columns = features
```

```python
# Đọc file subject train và subject test
subject_train = pd.read_csv('getdata_projectfiles_UCI HAR Dataset/UCI HAR Dataset/train/subject_train.txt', delim_whitespace=True, header=None)
subject_test = pd.read_csv('getdata_projectfiles_UCI HAR Dataset/UCI HAR Dataset/test/subject_test.txt', delim_whitespace=True, header=None)

# Ghép các dữ liệu lại với nhau, bao gồm subject, X và y
X_train_data = pd.concat([subject_train, X_train, y_train], axis=1)
X_train_data.columns = ['subject'] + features + ['activity']

print(X_train_data.head())
```

```python
X_test_data = pd.concat([subject_test, X_test, y_test], axis=1)
X_test_data.columns = ['subject'] + features + ['activity']
print(X_test_data.head())
```

```python
# Ghép dữ liệu train và test lại với nhau
X_data = pd.concat([X_train_data, X_test_data], axis=0)
print(X_data.shape)
```

## 4) Với các thuộc tính dữ liệu, chỉ lấy ra các thuộc tính giá trị trung bình (mean) và giá trị độ lệch chuẩn (standard deviation).


```python
# Lấy các cột có chứa mean() và std() trong tên cột
features = X_data.columns
mean_std_features = [feature for feature in features if 'mean()' in feature or 'std()' in feature or 'subject' in feature or 'activity' in feature]
feature_index = [X_data.columns.get_loc(feature) for feature in mean_std_features]
```

## 5) Đặt tên header cho bộ dữ liệu sao cho tên có ý nghĩa.


```python
# Thay đổi tên các cột trong X_data thành các tên dễ đọc hơn
import re 

rename_rules = {
    r'^t': 'Time domain:',
    r'^f': 'Frequency domain:',
    r'-': ',',
    r'mean\(\)': 'mean value',
    r'std\(\)': 'standard deviation value',
    r'-X': 'on X axis',
    r'-Y': 'on Y axis',
    r'-Z': 'on Z axis',
    r'AccJerk': 'acceleration jerk',
    r'Acc': 'acceleration',
    r'GyroJerk': 'angular velocity jerk',
    r'Gyro': 'angular velocity',
    r'Mag': 'magnitude'
}

def rename_column(col_name):
    for pattern, replacement in rename_rules.items():
        col_name = re.sub(pattern, replacement, col_name)
    return col_name

X_data.columns = [rename_column(col) for col in X_data.columns]
print(X_data.columns)
```

```python
X_data_filtered = X_data.iloc[:, feature_index]
print(X_data_filtered.head())
```

```python
# Tính trung bình của các cột theo từng activity và subject
X_data_filtered.groupby(['activity','subject']).mean().reset_index()
```

```python
# Đọc file activity_labels
activity_labels = pd.read_csv('getdata_projectfiles_UCI HAR Dataset/UCI HAR Dataset/activity_labels.txt', delim_whitespace=True, header=None)
activity_labels.columns = ['index', 'activity']

activity_labels = activity_labels.set_index('index')['activity'].to_dict()

# Thay thế các giá trị trong cột activity bằng tên hoạt động tương ứng
X_data_filtered.loc[:,'activity'] = X_data_filtered['activity'].map(activity_labels)

print(X_data_filtered.head())
```

Sau khi vẽ model thì ta có thể thấy có khá nhiều thuộc tính tương quan cao với nhau, ta có thể bỏ bớt các cột tương quan cao và giữ lại các cột tương quan thấp

```python
import pandas as pd
import numpy as np

# Tính ma trận tương quan
corr_matrix = X_data_filtered.iloc[:, 1:-1].corr().abs()

# Lấy tam giác trên của ma trận (vì ma trận này đối xứng)
upper = corr_matrix.where(np.triu(np.ones(corr_matrix.shape), k=1).astype(bool))

# Xác định các cột có tương quan cao hơn 0.9
to_drop = [column for column in upper.columns if any(upper[column] > 0.8)]

# Loại bỏ các cột đó khỏi dữ liệu
X_filtered_reduced = X_data_filtered.drop(columns=to_drop)

print("Các cột bị loại bỏ do tương quan cao:", to_drop)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Vẽ heatmap của ma trận tương quan
plt.figure(figsize=(20, 18))
sns.heatmap(X_filtered_reduced.iloc[:,1:-1].corr(), cmap='coolwarm', annot=False, fmt=".2f", linewidths=0.5)
plt.title("Heatmap of Correlation Matrix")
plt.show()
```

```python
# Xác định outliers và bỏ outliers 
Q1 = X_filtered_reduced.iloc[:, 1:-1].quantile(0.25)
Q3 = X_filtered_reduced.iloc[:, 1:-1].quantile(0.75)
IQR = Q3 - Q1

lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

filtered_data = X_filtered_reduced[~((X_filtered_reduced.iloc[:, 1:-1] < lower_bound) | (X_filtered_reduced.iloc[:, 1:-1] > upper_bound)).any(axis=1)]

print("Shape of data before removing outliers:", X_filtered_reduced.shape)
print("Shape of data after removing outliers:", filtered_data.shape)
```
